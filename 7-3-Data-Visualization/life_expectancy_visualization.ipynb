```python
import numpy as np
df_num = df.select_dtypes(include=[int, float, bool])
corr = df_num.corr()

# ìƒ‰ìƒ íŒ”ë ˆíŠ¸ ì„¤ì •: íŒŒëž€ìƒ‰(ì–‘ì˜ ìƒê´€), ë¹¨ê°„ìƒ‰(ìŒì˜ ìƒê´€)
cmap = sns.diverging_palette(10, 240, n=9, as_cmap=True)

# ë§ˆìŠ¤í‚¹ ë° í•˜ì–€ìƒ‰ìœ¼ë¡œ ì„¤ì •í•  ì˜ì—­ ì •ì˜
mask = np.zeros_like(corr, dtype=bool)
np.fill_diagonal(mask, True)

plt.figure(figsize=(15, 15))
sns.heatmap(corr, mask=mask, cmap=cmap, annot=True, fmt=".2f", center=0,
            linewidths=.5, cbar_kws={"shrink": .5})

# íŠ¹ìˆ˜í•œ ì²˜ë¦¬: ìƒê´€ ê³„ìˆ˜ê°€ 1ì¸ ê²½ìš° í•˜ì–€ìƒ‰ìœ¼ë¡œ í‘œì‹œ
for i in range(len(corr)):
    for j in range(len(corr)):
        if corr.iloc[i, j] == 1:
            plt.gca().add_patch(plt.Rectangle((j, i), 1, 1, fill=False, edgecolor='white', lw=3))

plt.title('Correlation Heatmap with Customized Colors')
plt.show()
```
